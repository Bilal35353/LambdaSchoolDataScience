# 3d plots

Use the following function `f(x,y)` and `np.meshgrid` to generate a beautiful 3d topology using [this tutorial](https://jakevdp.github.io/PythonDataScienceHandbook/04.12-three-dimensional-plotting.html).


```
import numpy as np

def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
```

* What shape are X,Y, and Z? Check out [meshgrid](https://docs.scipy.org/doc/numpy/reference/generated/numpy.meshgrid.html).
* How are you converting data in `x`, `y`, `f(x,y)`, `X`, `Y`, and `Z` into a pretty picture? In what way are all six data elements and the picture the same? How are they different?


```
print(X.shape, Y.shape, Z.shape)
```

`X`,  `Y`, and `Z` are all of shape `(40, 50)`. `x` can be thought of as a set of 50 points along the x axis, while `y` is a set of 40 points along the y axis. The set of all possible pairs of points from these two sets is then treated as the domain of the function `Z = f(x, y)`. These domain elements are captured by all the pairs of elements from `X` and `Y` that have corresponding indices, or positions within the matrices, whereas the element of `Z` with the same index is the point `Z = f(X=x, Y=y)`, or the result of applying the function to that x-y pair.

In order to make a grid, the rows of `X` are identical while the same is true of the columns of `Y`. This is done in order to ensure that all possible combinations of the elements of `x` and `y` are captured. The graph of the function, displayed below, plots all 2000 points corresponding to each unique index into `X`, `Z` and `Z`. It maps tthe value of each coordinate to the distance from its axis, and projects the final image onto a 2D surface. In order to produce a contour, surfaces are interpolated between the points provided by the three coordinate matrices.

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, 
                cmap='viridis', edgecolor=None)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show();
```

# Stretch goal

Make two more plots using data and `f(x,...)` of your own devising. Use $x^2$ if you don't have any other ideas. :)

```
# More 3d plots!

def g(x, y):
    return x*np.power(y, 3) - y*np.power(x, 3)

def h(x, y):
    return (-x)*y*np.exp(-np.power(x, 2)-np.power(y, 2))

def j(x, y):
    return x*np.power(y, 2)

def k(x, y):
    return x*y * np.sin(x)
```

```
for function in [g, h, j, k]:
    Z = function(X, Y)
    ax = plt.axes(projection='3d')
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1, 
                    cmap='viridis', edgecolor=None)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    plt.show();
```
